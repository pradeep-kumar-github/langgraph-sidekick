- Python REPL

an interactive command-line shell that lets you type Python code and see the results executed immediately.

How REPL Works Under the Hood:\
Read: The shell reads the Python expressions or statements you type.\
Evaluate: The Python interpreter runs your code to calculate a result.\
Print: The shell prints the resulting output directly to your screen.\
Loop: The environment loops back to the first step, waiting for your next command.
